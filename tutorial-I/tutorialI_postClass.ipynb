```python
num=1
while(num<=10):
    if(num%2==1):
        print(num)
    num=num+1
```

```python
num=1
sum=0
while(num<=10):
    print(num)
    sum=sum+num
    num=num+1
print("Sum is",sum)
```

```python
#Nested for loops
#print all prime numbers from 1-100
for num in range(1,101):
    isPrime=True
    for div in range(2,num-1):
        if(num%div):
            isPrime=False
    if(isPrime):
        print(num)
```
